## Procedimiento de medición

#### Método de medición:

Para realizar la medición del circuito, se utilizaron dos metodologías: En primer lugar, se realizaron un barrido y medición manual a cargo del grupo; y, una vez corroborado el correcto funcionamiento, se realizó un análisis automático utilizando un analizador de audio con asistencia docente.

#### Método #1

La metodología aplicada grupalmente consistió en introducir una señal, cuya frecuencia se varió dentro del espectro abarcado entre las frecuencias de corte, y evaluar la amplitud y desfasaje a la salida. Para esto, se contó con un generador de señales y un osciloscopio.


La amplitud de tensión se midió tanto en la entrada como en la salida, tomando el valor RMS de cada canal y ajustando la escala horizontal de forma tal que por lo menos 3 ciclos de la señal fueran visibles. La ganancia en dB se obtuvo a partir del siguiente cálculo:

$ |H_{dB}|=20·\log\left(\frac{v_i}{v_o}\right)$


Para medir el desfasaje, se colocaron dos cursores en picos adyacentes, uno de cada canal y se obtuvo la diferencia temporal entre ambos. El desfasaje se obtuvo a partir del siguiente cálculo:

$ \Delta \phi = 360^\circ · f_i · \Delta t_{ms}/ 1000 $


La variación de frecuencia se realizó aumentando la variación a medida que aumentaba la frecuencia, para mantener una resolución aproximadamente constante según el eje logarítmico horizontal.


Debido a que la medición de la señal filtrada se realizó con un osciloscopio, fue necesario tener en cuenta que el mismo puede presentar errores de resolución para señales de entrada de pequeña amplitud. Por este motivo, se aplicó una tensión de entrada de 10V, que luego de la atenuación esperada de 30dB produce una señal de 10mV, resoluble por el instrumento.


#### Método #2

El analizador de audio utilizado es capaz de producir el barrido de señal y almacenar los resultados de las mediciones correspondientes, de acuerdo a una serie de parámetros configurados para cada análisis. En este caso, configuramos las frecuencias de inicio y fin y la cantidad de puntos (100).

### Resultados de la medición a través del método #1

Los datos correspondientes a la medición se ecuentran en el <a href="./mediciones_tplab.csv">siguiente archivo </a>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('./mediciones_tplab.csv')

ganancia = np.array(data["Ganancia [dB]"])
fase = np.array(data["Fase [°]"])
frecuencia = np.array(data["f [Hz]"],dtype=int)

ganancia = [float(a.replace(',','.')) for a in ganancia]
fase = [float(a.replace(',','.')) for a in fase]

plt.figure(1)
plt.semilogx(frecuencia,ganancia)
plt.title('Ganancia vs Frecuencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Ganancia [dB]')

plt.figure(2)
plt.semilogx(frecuencia,fase)
plt.title('Fase vs frecuencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [°]')
```

### Resultados de la medición a través del método #2

Los datos correspondientes a la medición de la ganancia se encuentran en el <a href="./Sweep Data_0_modulo.csv">siguiente archivo</a>

Los datos correspondientes a la medición de la fase se encuentran en el <a href="./Sweep Data_1_fase.csv">siguiente archivo</a>

```python
data_g = pd.read_csv("./Sweep Data_0_modulo.csv")
data_f = pd.read_csv("./Sweep Data_1_fase.csv")

frec_g = np.array(data_g["X (Hz)"],dtype='float')
gain_g = np.array(data_g["Ch-1 (dBr)"],dtype='float')

frec_f = np.array(data_f["X (Hz)"],dtype='float')
fase_f = np.array(data_f["Ch-1 (deg)"],dtype='float')

fase_f = [fase_f[i] + 360 * (fase_f[i] < 0) for i in range(0,len(fase_f))]

plt.figure(3)
plt.semilogx(frec_g,gain_g)
plt.title('Ganancia vs Frecuencia')
plt.ylabel('Ganancia [dB]')
plt.xlabel('Frecuencia [Hz]')

plt.figure(4)
plt.semilogx(frec_f,fase_f)
plt.title('Fase vs frecuencia')
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Fase [°]')
```
