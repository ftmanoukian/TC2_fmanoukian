## Simulación numérica

```python
import numpy as np
from pytc2.sistemas_lineales import pzmap, bodePlot, TransferFunction

a = 0.6264564863
b = a
c = 1.14244773
Q = 5

num_bp = np.polymul([a/Q,0],[c/(Q**2),0,0])
den_bp = np.polymul([
    1,a/Q,1
],[
    1,b/Q,2+c/(Q**2),b/Q,1
])

tf_bp = TransferFunction(num_bp,den_bp)
pzmap(tf_bp)
bodePlot(tf_bp)
```
