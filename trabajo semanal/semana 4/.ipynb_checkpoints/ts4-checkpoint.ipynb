## Simulación numérica:

```python
from pytc2.sistemas_lineales import TransferFunction, pzmap, bodePlot
import numpy as np


alpha_max = 1
eps = np.sqrt(10**(alpha_max/10)-1)
N = 3
wb = eps**(-1/N)

#polo simple
num1 = [0,1,0]
den1 = [0,1,1/wb]

#polos complejos conjugados
num2 = [1,0,0]
den2 = [1,1/wb,1/wb**2]

#transferencia normalizada completa:
num = np.polymul(num1,num2)
den = np.polymul(den1,den2)
T = TransferFunction(num,den)

pzmap(T)
bodePlot(T)
```
